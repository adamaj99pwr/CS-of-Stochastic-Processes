$e_i \sim Exp(1)$ i.i.d $\Rightarrow \Gamma \sim \Gamma$ not i.i.d

$\sum_{i}^{N} \epsilon_i \Gamma^{-1/\alpha} \Rightarrow S_{\alpha}(c_{\alpha}^{-1/\alpha}, \beta=0, \mu=0)$,
$c_{\alpha} = \frac{2\sin \frac{\pi \alpha}{2} \Gamma (\alpha)}{\pi}$

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.distributions.empirical_distribution import ECDF
from scipy.stats import norm, kstest, levy_stable, cauchy
```

```python
alpha = 1.6
```

```python
alphas = []

for i in range(1000):
    X = []
    for j in range(10000):
        epsilon = np.random.choice([-1,1], p=[0.5, 0.5], size=100)
        e = np.random.exponential(1, size=100)
        gamma = np.cumsum(e)
        X.append(np.sum(epsilon * gamma ** (1/alpha)))
        
    a, b, g, d = levy_stable._fitstart(X)
    alphas.append(a)
```
