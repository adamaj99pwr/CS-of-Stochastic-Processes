```python
import numpy as np
from scipy.stats import norm, spearmanr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [9, 5]
plt.rcParams['figure.dpi'] = 110
```

```python
def unif(R, n):
    k = np.shape(R)[0]
    C = 2*np.sin(np.pi/6 * R)
    z = np.random.normal(0, 1, size=(n, k))
    chol = np.linalg.cholesky(C)
    u = pd.DataFrame(norm.cdf(z @ chol))
    u.columns = ['x', 'y']
    return u
```

```python
def fackler(alpha, beta, lambd, R, n):
    u = unif(R, n)
    u1 = u['x']
    u2 = u['y']
    X = -1/beta * np.log(u1)
    Y = lambd * (u2**(-1/alpha) - 1)
    return X, Y
```

```python
R1 = np.array([[1,0], [0,1]])
u1 = unif(R1, 4)
```

```python
x1, y1 = fackler(1.8, 1, 1, R1, 10000)
```

```python
R2 = np.array([[1, 0.6], [0.6, 1]])
u2 = unif(R2, 4)
```

```python
x2, y2 = fackler(1.8, 1, 1, R2, 10000)
```

```python
R3 = np.array([[1, -0.6], [-0.6, 1]])
u3 = unif(R3, 4)
```

```python
x3, y3 = fackler(1.8, 1, 1, R3, 10000)
```
