```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable, norm
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = [9, 5]
plt.rcParams['figure.dpi'] = 120
```

```python
def levy_stable_motion(T, I, alpha, beta):
    tau = T / I
    ts = np.arange(0, T, tau)
    ksi = levy_stable.rvs(alpha, beta, loc=0, scale=tau**(1/alpha), size=I-1)
    X = np.zeros(I)
    X[1:] = np.cumsum(ksi)
    return X 
```

```python
def plot_levy_motion(N, T, I, alpha, beta):
    xs = levy_stable_motion(T, I, alpha, beta)
    for i in range(N-1):
        xs_new = levy_stable_motion(T, I, alpha, beta)
        xs = np.vstack((xs, xs_new))
    df = pd.DataFrame(xs.T)
    df.plot(legend=False, title=rf'Levy stable motion for $\alpha$={alpha}, $\beta$={beta}, T={T}, I={I}', color='black')
    plt.show()
```

```python
def plot_levy_motion2(N, T, I, alpha, beta):
    xs = levy_stable_motion(T, I, alpha, beta)
    for i in range(N-1):
        xs_new = levy_stable_motion(T, I, alpha, beta)
        xs = np.vstack((xs, xs_new))
    df = pd.DataFrame(xs.T)
    return df
```

```python
def plot_levy_increments(N, T, I, alpha, beta):
    tau = T / I
    xs = levy_stable.rvs(alpha, beta, loc=0, scale=tau**(1/alpha), size=I-1)
    for i in range(N-1):
        xs_new = levy_stable.rvs(alpha, beta, loc=0, scale=tau**(1/alpha), size=I-1)
        xs = np.vstack((xs, xs_new))
    df = pd.DataFrame(np.transpose(xs))
    df.plot(legend=False, title=rf'Levy stable motion increments for $\alpha$={alpha}, $\beta$={beta}, T={T}, I={I}')
    plt.show()
```

```python
def quantile_lines(N, T, n, alpha, beta, qs):
    m = len(qs)
    xs_matrix = np.zeros((n, N))
    quantile_matrix = np.zeros((n, m))
    for i in range(N):
        xs_matrix[:,i] = levy_stable_motion(T, n, alpha, beta)
    for i in range(n):
        xs = xs_matrix[i]
        xs_sorted = sorted(xs)
        quantile_matrix[i] = [xs_sorted[int(q*N)] for q in qs]
    quantile_df = pd.DataFrame(quantile_matrix, columns=[f'Q{round(q, 2)}' for q in qs])
    return quantile_df
```

```python
def norm_quant(qs, ts):
    qlines = np.zeros((len(ts), len(qs)))
    for i in range(len(qs)):
        qlines[:,i] = np.sqrt(ts)*qs[i] # norm.ppf(qs[i], loc=0, scale=2)
    return pd.DataFrame(qlines, columns=[f'Q{round(q, 2)}' for q in qs])
```

```python
ql = quantile_lines(1000, 10, 1000, 2, 1, np.arange(0.1, 1, 0.1))
lm = plot_levy_motion2(10, 10, 1000, 2, 1)
```

```python
nq = norm_quant(np.arange(0.1, 1, 0.1), np.linspace(0, 10, 1000))
nq.plot()
```

```python
ax = lm.plot(legend=False, title=r'Levy stable motion for $\alpha$=2, $\beta$=1, T=10, I=1000', color='black', linewidth=0.5)
ql.plot(ax=ax, linewidth=2)
xticks = list(range(0, 1001, 100))
xtick_labels = np.linspace(0, 10, 11)
ax.set_xticks(xticks, xtick_labels)
plt.grid()
plt.show()
```
